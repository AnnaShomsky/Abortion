```python
health_abortion_legal2= health_abortion_legal.drop_duplicates(subset=["Country"], keep='first', ignore_index=True)
health_abortion_legal2["Country"].count()
```

```python
health_abortion_legal2["abortion_code"]= np.where(health_abortion_legal2["legality"] == "No restriction", 1, 
                                                 np.where(health_abortion_legal2["legality"] == "Prohibited altogether",3,2))
                                                 
                                                 
```

```python
abortion_numeric= health_abortion_legal2[["Population size in millions", "Health spending per capita", "Health spending as percent of GDP", "Life expectancy in years", "The number of crude births per 1000 people per year", "Fertility rate births per woman","Prevalence of HIV percent of the population ages 15-49", "Suicide mortality rate per 100000 population","Maternal mortality per 100000 live births","Deaths of children under five years of age per 1000 live births","Neonates dying before 28 days of age per 1000 births","Infant deaths per 1000 live births","Deaths of children under five years of age per 1000 live births","Public spending on education percent of GDP","Cost of living index world average = 100","Happiness Index 0 (unhappy) - 10 (happy)",
                                          'Human Development Index (0 - 1)','Women in parliament percent','numberofabortions','abortionrate','perc_p_ending_in_a', 'abortion_code' ]]

abortion_numeric.head()
```

```python
df_pearson = abortion_numeric.corr(method='pearson')
sns.set(rc = {'figure.figsize':(20, 12)})
go=sns.heatmap(df_pearson, annot=True, cmap="YlGnBu")

ax = go
```
